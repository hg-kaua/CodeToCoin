for support levels we are going to look for a certain number of decreasing lows that are preceding our candle of interest and the candle should have a lowe low value than the other candles. Then we are going to have three increasing lows after the candle of interest

same thing for resistance levels we have an increasing highs then higher high of the candle of interest then decreasing highs after the candle of interest

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import datetime as dt 
import yfinance as yf



yf.pdr_override()
start = dt.datetime(2015,1,1)
end = dt.datetime.now()
ticker = 'TAEE11.SA'
df_data = pdr.get_data_yahoo(ticker, start, end)

df_data
```

```python
def support(df, l, n1, n2): # n1 and n2 are the number of candles after candle l
    
    for i in range(l - n1 + 1, l + 1):
        if(df['Low'].iloc[i] > df['Low'].iloc[i - 1]):
            return 0
    
    for i in range(l + 1, l + n2 + 1):
        if (df['Low'].iloc[i] < df['Low'].iloc[i - 1]):
            return 0
    
    return 1

support(df_data, 46, 3, 2)
```

```python
def resistance(df, l, n1, n2): # n1 and n2 are the number of candles after candle l
    
    for i in range(l - n1 + 1, l + 1):
        if(df['High'].iloc[i] < df['High'].iloc[i - 1]):
            return 0
    
    for i in range(l + 1, l + n2 + 1):
        if (df['High'].iloc[i] > df['High'].iloc[i - 1]):
            return 0
    
    return 1

resistance(df_data, 46, 3, 2)
```

```python
sr = []
n1 = 2
n2 = 1

for row in range(3, 200):
    if support(df_data, row, n1, n2):
        sr.append((row, df_data["Low"].iloc[row], 1))
    if resistance(df_data, row, n1, n2):
        sr.append((row, df_data['High'].iloc[row], 2))
        
print(sr)
```

```python
s = 0
e = 200
dfpl = df_data[s:e]
import plotly.graph_objects as go
from datetime import datetime
import matplotlib.pyplot as plt

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                open=dfpl['Open'],
                high=dfpl['High'],
                low=dfpl['Low'],
                close=dfpl['Close'])])

c=0
while (1):
    if(c>len(sr)-1 ):#or sr[c][0]>e
        break
    fig.add_shape(type='line', x0=s, y0=sr[c][1],
                  x1=e,
                  y1=sr[c][1]
                  )#x0=sr[c][0]-5 x1=sr[c][0]+5
    c+=1
fig.show()
```
