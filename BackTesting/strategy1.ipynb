First strategy:

- calculate the mean of candles size, in order to find which candle is bigger than mean of candles

- calculate the mean of volumes bar size, in order to find which volume bar is bigger than mean of volumes bar

- check if after this candle the price market became lower or higher

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt




# fetch historical data
start = dt.datetime(2015,1,1)
end = dt.datetime.now()
stock_data = yf.download('TAEE11.SA', start, end, interval='1d')

stock_data
```

```python
stock_data["Size Candle"] = np.abs(stock_data["Close"] - stock_data["Open"])

stock_data.tail(20)
```

```python
# check candle mean
stock_data["Mean Size Candle"] = stock_data['Size Candle'].rolling(100).quantile(0.9)
# check volume mean
stock_data["Mean Size Volume"] = stock_data['Volume'].rolling(100).quantile(0.9)
# check sign
stock_data["Sign"] = np.sign(stock_data["Close"] - stock_data["Open"])

stock_data.tail(20)
```

```python
cond1 = stock_data["Size Candle"] > stock_data["Mean Size Candle"]
cond2 = stock_data["Volume"] > stock_data["Mean Size Volume"]
cond3 = stock_data["Sign"] == -1

conditions_met = stock_data[cond1 & cond2 & cond3]
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1, shared_xaxes=True, specs=[[{"secondary_y": True}],[{}]],
                    vertical_spacing=0.05, subplot_titles=('OHLC', 'Volume'), row_width=[0.2, 0.8])
# Add candlestick trace
fig.add_trace(go.Candlestick(x=stock_data.index,
                             open=stock_data["Open"],
                             high=stock_data["High"],
                             low=stock_data["Low"],
                             close=stock_data["Close"]), row=1, col=1)

# Add scatter trace to highlight the entry points
fig.add_trace(go.Scatter(x=conditions_met.index, y=conditions_met["Close"], mode='markers', 
                         marker=dict(size=10, color='black', symbol='triangle-up'), name='Entry Points'), row=1, col=1)

# Add volume bar trace
fig.add_trace(go.Bar(x=stock_data.index, y=stock_data["Volume"], name='Volume'), row=2, col=1)

# Add mean volume trace
fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data["Mean Size Volume"], mode='lines', name='Mean Size Volume'), row=2, col=1)

# Update layout
fig.update_layout(height=700, xaxis_rangeslider_visible=False, template='seaborn')
fig.update_layout(margin=dict(l=30, r=30, t=30, b=30))

# Show the plot
fig.show()
```
