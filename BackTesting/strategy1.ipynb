First strategy:

- calculate the mean of candles size, in order to find which candle is bigger than mean of candles

- calculate the mean of volumes bar size, in order to find which volume bar is bigger than mean of volumes bar

- check if after this candle the price market became lower or higher

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt




# fetch historical data
start = dt.datetime(2015,1,1)
end = dt.datetime.now()
stock_data = yf.download('TAEE11.SA', start, end, interval='1d')

stock_data
```

```python
stock_data["Size Candle"] = np.abs(stock_data["Close"] - stock_data["Open"])

stock_data.tail(20)
```

```python
# check candle mean
stock_data["Mean Size Candle"] = stock_data['Size Candle'].rolling(100).quantile(0.9)
# check volume mean
stock_data["Mean Size Volume"] = stock_data['Volume'].rolling(100).quantile(0.9)
# check sign
stock_data["Sign"] = np.sign(stock_data["Close"] - stock_data["Open"])

stock_data.tail(20)
```

```python
cond1 = stock_data["Size Candle"] > stock_data["Mean Size Candle"]
cond2 = stock_data["Volume"] > stock_data["Mean Size Volume"]
cond3 = stock_data["Sign"] == -1

conditions_met = stock_data[cond1 & cond2 & cond3]
```
