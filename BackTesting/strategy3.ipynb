```python
start = dt.datetime(2015,1,1)
end = dt.datetime.now()
ticker = 'BBAS3.SA'

df_data = pdr.get_data_yahoo(ticker, start, end)

df_data
```

```python
# identify the triple bottom Pattern

# to identify the triple bottom, we need to look for three roughly equal lows 
# followed by a breakout above the resistance level.

def identify_triple_bottom(df, window=20, tolerance=0.01):
    df['Triple_Bottom'] = np.nan
    
    for i in range(window, len(df) - window):
        low1 = df['Low'][i - window:i].min()
        low2 = df['Low'][i:i + window].min()
        low3 = df['Low'][i + window:i + 2 * window].min()
        
        if (abs(low1 - low2)/ low1 <= tolerance) and (abs(low2 - low3) / low2 < tolerance):
            # Check for breakout above the resistance level
            resistance = df['High'][i - window:i].max()
            if df['Close'][i + 2 * window] > resistance:
                df['Triple_Bottom'][i + 2 * window] = df['Close'][i + 2 * window]
        
    return df

df_data = identify_triple_bottom(df_data)
```
